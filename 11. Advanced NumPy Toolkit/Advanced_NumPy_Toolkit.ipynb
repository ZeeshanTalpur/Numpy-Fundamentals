# <center>Advanced NumPy Toolkit (Data Cleaning & Logic)</center>

**In real datasets, the problem is not:**\
How do i calculate means?\
**The problem is often:**\
What do I do with missing values?\
How do I filter bad data?\
How do I extract only meaningful rows?\
How do I handle conditions efficiently?\
**This toolkit solves these very problems**

**1) Boolean Masking(Conditional Selection)**

```python
import numpy as np
data = np.array([10,25,30,45,50])
print(f"Boolean Mask: {data>30}")
```

Boolean Mask can also be used to print those elements of an array where condition is satisfied.

```python
data[data>30]
```

**2) Multiple Conditions**\
Always keep each condition within parentheses\
Use '&' not "and"\
Use '|' not "or"\
Use '~' not "not"

**a. AND:**

```python
data[(data>20) & (data<50)]
```

**b. OR:**

```python
data[(data>20) | (data<50)]
```

**c. NOT:**

```python
data[~(data>30)]
```

**3) np.where() (Decision Engine)**\
It's like if-else for arrays

```python
ages = np.array([17,14,21,12,23,62])
np.where(ages>=18,"Adult","child")
```

**4) np.select() (Multi Condition Logic)**\
useful when there are multiple conditions

```python
conditions=[
    ages<18,
    (ages>=18) & (ages<=60),
    ages>60
]
choices = ["child","adult","retired"]
np.select(conditions,choices,default="")
```

**5) Handling Missing Data (NaN)**

a. Create missing values:

```python
data = np.array([10,np.nan,30,np.nan,50])
```

b. Detect missing values:

```python
np.isnan(data)
```

c. Remove missing values:

```python
data[~np.isnan(data)]
```

**6) NaN safe statistics:**\
Normal functions break with NaN\
np.mean(data)  #may return NaN\
NaN safe methods are used as follows:

```python
print(np.nanmean(data))
print(np.nanmedian(data))
print(np.nanstd(data))
```

**7) np.clip() (Value Control)**\
Used for limiting values

```python
data = np.array([10,20,30,40,50])
np.clip(data,20,40)
```

**8) argmax & argmin**\
argmax: index of max value in the array\
argmin: index of min value in the array

```python
data = np.array([10,50,30])
print(np.argmax(data))
print(np.argmin(data))
```

**9) argsort()**\
sort but return indices only.

```python
data = np.array([40,10,30])
np.argsort(data)
```

**10) np.unique()**\
Find distinct values

```python
data = np.array([1,1,2,2,3,3,3])
np.unique(data)
```

With counts:

```python
np.unique(data,return_counts=True)
```

**11) Logical Reductions(any & all)**

**a. any():** True if any value matches condition\
**b. all():** True if all values match condition

```python
data = np.array([10,20,30,40,50])
print(np.any(data>40))
print(np.all(data>0))
```

**REAL INDUSTRY EXAMPLE**(Everything Combined)

```python
data = np.array([10,25,np.nan,45,60,np.nan])
```

**Step1**: Clean data

```python
clean = data[~np.isnan(data)]
```

**Step2**: Clip Extreme Values

```python
clean = np.clip(clean,20,50)
```

**Step3**: Label data

```python
np.where(clean>40,"High","Low")
```

**Step4**: NaN Safe Statistics

```python
np.nanmean(clean)
```

```python
np.nanstd(clean)
```
