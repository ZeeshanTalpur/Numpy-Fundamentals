**<center>Broadcasting</center>**\
NumPy's ability to:\
Automatically expand smaller arrays so arithmetic can be performed with larger arrays.

**Broadcasting Rule # 1**\
Start comparing dimensions from the RIGHT.\
**Broadcasting Rule #2**\
Dimensions are compatible if: Equal OR One of them is 1(Because 1 can be expanded to any number with repititions to match the dimensions)

**a) Scalar Broadcasting:** Scalar is expanded in an array to match dimensions of the array it is supposed to be added with

```python
import numpy as np
arr = np.array([
    [1,2,3],
    [4,5,6]
])
arr + 5  #5 is the scalar
```

In the above example, the array's shape is (2,3). Hence, the scalar 5 is expanded to be contained in an array of shape(2,3)\
5 becomes\
[\
[5 5 5],\
[5 5 5]\
]\
Then, this is added with the given array which yields the output.

**b) Row Broadcasting:** Row vector is repeated downwards to match the dimensions of the array it is supposed to be added with

```python
arr = np.array([
    [1,2,3],
    [4,5,6]
])
row = np.array([10,20,30])
arr + row
```

In the above example, the array's shape is (2,3), and the row vector has a shape of (3,) or simply (1,3). Hence the **row** needs to be expanded to 2.\
[ 10 20 30 ]\
becomes\
[\
[10 20 30]\
[10 20 30]\
]\
Then, it is added with the given array which yields the output

**c) Column Broadcasting:** Column vector is repeated sideways to match the dimensions of the array it is supposed to be added with

```python
arr = np.array([
    [1,2,3],
    [4,5,6]
])
column = np.array([
    [10],
    [20]
])
arr + column
```

In the above example, the array's shape is (2,3), and the column vector has a shape of (2,1). Hence the **column** needs to be expanded to 3.\
[\
[ 10 ]\
[ 20 ]\
]\
becomes\
[\
[10 10 10]\
[20 20 20]\
]\
Then, it is added with the given array which yields the output

**Real Data Science Example 1**\
**Suppose:**\
temperatures = np.array([\
    [25,26,24],\
    [30,31,29],\
    [20,22,21]\
])\
**Rows:** Cities\
**Columns:** Days\
Suppose each city's thermometer is off by:\
correction = np.array([\
    [1],\
    [2],\
    [1]\
])\
**Add:**\
temperatures + correction\
Each city receives its own adjustment.

```python
temperatures = np.array([
[25,26,24],
[30,31,29],
[20,22,21]
])
correction = np.array([
[1],
[2],
[1]
])
temperatures + correction
```

**Real Data Science Example 2**\
**Feature Scaling**\
**Suppose:**\
X = np.array([\
    [22,50000],\
    [25,70000],\
    [30,90000]\
])\
**Columns:** Age | Salary\
**Mean values:**\
means = np.array([25,70000])\
**Subtract:** X - means\
Broadcasting automatically applies the correct mean to each column.\
This is done constantly in machine learning.

```python
X = np.array([
[22,50000],
[25,70000],
[30,90000]
])
means = np.array([25,70000])
X - means
```

# <center>Professional Uses of Broadcasting</center>
You'll use broadcasting for:\

X - mean\

(X - mean) / std\
**3. Image Processing**\
image * brightness_factor\
**4. Sensor Calibration**\
sensor_data + correction\
**5. Financial Modeling**\
revenue * tax_rates\
**6. Machine Learning**\
X @ W + b\
That + b works because of broadcasting.
