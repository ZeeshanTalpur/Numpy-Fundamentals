Indexing and Slicing allow us to extract specificc pieces of data from a larger dataset.

**INDEXING:**

```python
import numpy as np
arr = np.array([10,20,30,40,50])
```

**Positive Indexing:**
Starts from left to right

```python
print(arr[2])
```

**Negative Indexing:**
Starts from right to left

```python
print(arr[-2])
```

**SLICING:**

```python
print(arr[1:3])
```

**General Form:**
arr[ start : stop : step]

a. Start\
Where to begin

```python
print(arr[2:])  #Start at index 2
```

b. Stop\
Where to stop

```python
print(arr[:3])  #Stop at index 3
```

c. Step\
How many positions to jump

```python
print(arr[::2])
```

**Reversing an Array:**\
One of the coolest use of slicing

```python
print(arr[::-1])
```

**Why Do We Slice Rows and Columns?**\
In Data Science:\
Rows = observations\
Columns = features

**Example:**

| Name  | Age | Salary |
| ----- | --- | ------ |
| Ali   | 22  | 50000  |
| Sara  | 25  | 70000  |
| Ahmed | 30  | 90000  |


Lets make an array for the table above

```python
arr = np.array([
    [22, 50000],
    [25, 70000],
    [30, 90000]
])
print(arr)
```

Sometimes we need only ages - Let's slice only ages

```python
print(arr[:,0])
```

Sometimes we only need salaries - Let's slice only salaries

```python
print(arr[:,1])
```

Sometimes we need first few records - Let's slice 2 records 

```python
print(arr[:2,])
```

Sometimes we need only ages of first two records - Let's slice ages of first two records

```python
print(arr[:2,0])
```

Sometimes we need only salaries of first two records - Let's slice salaries of first two records

```python
print(arr[:2,1])
```

**Boolean Indexing - The Heart of Data Science**\
Now, we move beyond positions\
Instead of saying give me index 5\
We say, give me everyone older than 18

```python
ages=np.array([17,22,23,21,19,14])
ages>=18
```

Above is a Boolean Mask - very useful because it serves as a filter and unlike lists, it omits out the need for loops to validate

**Fancy Slicing**\
Normal slicing takes a range\
Fancy slicing lets you choose specific positions

```python
print(arr[[1,2]]) #Fancy indexing rows
```

```python
print(arr[:,[0]]) #Fancy indexing columns
```
