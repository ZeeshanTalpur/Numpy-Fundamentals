**4. Reproducibility**\
**Suppose:**\
np.random.randint(1,10,5)\
Run again:\
Different result.\
Run again:\
Different result.\
This is bad for experiments.\
**Use:**\
np.random.seed(42)\
**Then:**\
np.random.randint(1,10,5)\
will always generate the same numbers.

```python
np.random.seed(10)
np.random.randint(1,10,4)
```

Try running above block again and again, You'll get the same output everytime

**Why Data Scientists Care**\
**Suppose:**\
Train Accuracy = 92%\
Your teammate should be able to reproduce your results.\
Seeds help achieve this.

**6. Shuffling**\
Mixes elements of an array and switches their positions

```python
arr = np.array([1,2,3,4,5,6])
np.random.shuffle(arr)
print(arr)
```

Important: shuffle() modifies original array.

**7. Permutation**\
Similar to shuffle but returns a new array.\
Original remains unchanged

```python
arr = np.array([1,2,3,4,5,6])
np.random.permutation(arr)
```

```python
print(f"Original Unchanged Array: {arr}")
```

**10. Random Data Sampling:**


```python
population = np.arange(10000)
sample = np.random.choice(
    population,
    size=500,
    replace=False
)
print(sample)
```

**11. Simple Probability Simulation**\
Let's estimate probability

```python
tosses = np.random.choice(
    ["H","T"],
    size=10000
)
#np.sum(tosses=="H")
np.mean(tosses=="H")
```

**Why This Is Powerful**\
Instead of solving mathematically:\
P(event)\
you can simulate millions of experiments.\
This is called:**Monte Carlo Simulation**

**12. Monte Carlo Example**\
Estimate probability of rolling a 6.

```python
dice = np.random.randint(1,7,100000)
np.mean(dice==6)
```

**Industry Mini Project**\
**Create:**\
1000 customers\
**with:**\
Random age\
Random salary\
Random spending score\
**Then:**\
Compute means.\
Compute stds.\
Find correlation.\
Sample 100 customers.\
Shuffle data.\
Find probability salary > 80,000.\
Find probability age < 25.\
This resembles the first step of many real-world analytics projects.

```python
age = np.random.randint(16,60,1000)
salary = np.random.normal(50000,3000,1000)
spending = np.random.randint(1,6,1000)
customers = np.column_stack((age,salary,spending))
print(f"Means of age, salary and spending score: {np.mean(customers,axis=0)}")
print(f"Standard Deviations of age, salary and spending score: {np.std(customers,axis=0)}")
print(f"Correlation Matrix: \n{np.corrcoef(customers,rowvar=False)}")
indices = np.random.choice(500, size=100, replace=False)
sample = customers[indices]
np.set_printoptions(suppress=True)
print(f"Sample of 100 Customers:\n{sample}")
np.random.shuffle(sample)
print(f"Shuffled Sample:\n{sample}")
prob_salary_above_80000 = np.mean(salary > 80000)
print(f"Probability of salary > 80,000: {prob_salary_above_80000}")
prob_age_below_25 = np.mean(age < 25)
print(f"Probability of age below 25: {prob_age_below_25}")
```
