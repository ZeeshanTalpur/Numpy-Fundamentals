**Problem 11: Salary Analytics**\
salary = np.array([\
    45000,\
    50000,\
    52000,\
    55000,\
    58000,\
    60000,\
    62000,\
    250000\
])\
**Tasks:**\
1.	Compute mean salary.\
2.	Compute median salary.\
3.	Compute Q1, Q2, Q3.\
4.	Compute IQR.\
5.	Detect outliers using IQR rule.

```python
salary = np.array([45000,50000,52000,55000,58000,60000,62000,250000])
print(f"Mean Salary: {np.mean(salary)}")
print(f"Median Salary: {np.median(salary)}")
q1 = np.percentile(salary,25)
q2 = np.percentile(salary,50)
q3 = np.percentile(salary,75)
print(f"Quartiles:\nQuartile 1: {q1}\nQuartile 2: {q2}\nQuartile 3: {q3}")
iqr = q3 - q1
print(f"Interquartile Range(IQR): {iqr}")
outliers = salary[(salary<q1-1.5*iqr) | (salary>q2+1.5*iqr)]
print(f"Outliers: {outliers}")
```

**Problem 13: Correlation Study**\
hours = np.array([1,2,3,4,5,6])\
scores = np.array([40,50,60,70,80,90])\
**Tasks:**\
1.	Compute correlation matrix.\
2.	Extract actual correlation coefficient.\
3.	Interpret result.\
4.	Compute covariance matrix.

```python
hours = np.array([1,2,3,4,5,6])
scores = np.array([40,50,60,70,80,90])
correlation = np.corrcoef(hours,scores)
print(f"Correlation Matrix: \n{correlation}")
print(f"Actual Correlation Coefficient: {correlation[0,1]}")
print(f"It's a perfect positive correlation")
covariance = np.cov(hours,scores)
print(f"Covariance Matrix: \n{covariance}")
```

**Problem 15: Data Transformation Pipeline**\
data = np.arange(1,25)\
Tasks:\
1.	Reshape into 4×6.\
2.	Transpose it.\
3.	Flatten it.\
4.	Reshape into 3×8.\
5.	Verify size remains unchanged throughout.

```python
data = np.arange(1,25)
print(f"Original Array: {data}")
reshaped = data.reshape(4,-1)
print(f"Reshaped into 4 X 6:\n{reshaped}\nSize: {reshaped.size}")
transpose = reshaped.T
print(f"Transposed Array:\n{transpose}\nSize: {transpose.size}")
flattened = transpose.flatten()
print(f"Flattened Array: {flattened}\nSize: {flattened.size}")
again_reshaped = flattened.reshape(3,-1)
print(f"Reshaped into 3 X 8: \n{again_reshaped}\nSize: {again_reshaped.size}")
```
