**Problem 20: Mini Data Science Case Study**\
A company dataset:\
data = np.array([\
       [101,22,50000],\
       [102,35,120000],\
       [103,19,25000],\
       [104,41,90000],\
       [105,28,70000],\
       [106,17,15000]\
])\
**Columns:**\
EmployeeID | Age | Salary\
**Without loops, solve:**\
Extract all adults.\
Extract employees earning > 60000.\
Extract employees aged between 20 and 40.\
Extract Employee IDs of high earners.\
Extract the Age and Salary columns only.\
Extract first 3 records.\
Extract last 2 records.\
Extract every alternate record.\
Reverse the entire dataset.\
Verify shape, ndim, size and dtype.

```python
data = np.array([
[101,22,50000],
[102,35,120000],
[103,19,25000],
[104,41,90000],
[105,28,70000],
[106,17,15000]
])
adults = data[data[:,1]>=18]
print("Adults: \n",adults)
employees_earning_greater_than_60k = data[data[:,2]>60000]
print("Employees earning greater than 60k: \n",employees_earning_greater_than_60k)
employees_bw_20_and_40 = data[(data[:,1]>20) & (data[:,1]<40)]
print("Employees between ages 20 and 40: \n",employees_bw_20_and_40)
ids_of_high_earners = employees_earning_greater_than_60k[:,0]
print("IDs of high earners: ", ids_of_high_earners)
age_salary = data[:,1:]
print("Ages and Salaries only: \n",age_salary)
first_3records = data[:3]
print("First three records: \n",first_3records)
last_2records = data[-2:]
print("Last two records: \n", last_2records)
alternate_records = data[::2]
print("Alternate records: \n",alternate_records)
reversed_data = data[::-1]
print("Reversed dataset: \n", reversed_data)
print("Shape: ",data.shape)
print("Number of Dimensions: ",data.ndim)
print("Size: ",data.size)
print("DataType: ",data.dtype)
```
