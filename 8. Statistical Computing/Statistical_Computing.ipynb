**<center>STATISTICAL COMPUTING</center>**

Looking at raw numbers tells you very little. Statistics summarizes data into meaningful information.\
**Statistical Measures:**
 1) **Mean:** Central Value of the dataset

```python
import numpy as np
arr = np.array([52,55,56,57])
print(np.mean(arr))
```

2) **Median**: Middle value after sorting\
It is less sensitive to outliers unlike mean

```python
arr = np.array([10,20,30,40,500])
print(np.median(arr))
```

**Mean vs Median** (in terms of sensitivity to outliers) 

```python
print(f"Mean: {np.mean(arr)}")
print(f"Median: {np.median(arr)}")
```

In the above example, median better represents the typical value as compared to mean because it's less sensitive to the outliers.

3) **Minimum and Maximum**: least and greatest value

```python
print(f"Maximum value: {np.max(arr)}") 
print(f"Minimum value: {np.min(arr)}")
```

There's another way to do this --->

```python
print(f"Maximum value: {arr.max()}") 
print(f"Minimum value: {arr.min()}")
```

4) **Range**: Maximum - Minimum

```python
print(f"Range: {arr.max() - arr.min()}")
```

5) **Variance**: How spread out the data is\
$\operatorname{Var}(X) = \frac{1}{n}\sum_{i=1}^{n}(x_i - \bar{x})^2$

```python
print(f"Variance: {np.var(arr)}")
```

The purpose of squaring the differences is to prevent cancellation

6) **Standard Deviation**: Also tells how spread out the data is, but is easier to interpret\
      $\sigma = \sqrt{\operatorname{Var}(X)}$

```python
print(f"Standard Deviation: {np.std(arr)}")
```

7) **Percentiles**: indicates the value below which a given percentage of observations in a dataset falls

```python
results = np.array([56,65,43,78,92,65,55,57,59])
print(f"50th Percentile: {np.percentile(results,50)}")
```

Above example's output implies that 50% of the students scored less than 59 score and rest 50% above it

8) **Quartiles**: some special percentiles\
   1st Quartile --> 25th Percentile\
   2nd Quartile --> 50th Percentile\
   3rd Quartile --> 75th Percentile\
   In general, nth Quartile refers (n/4)th part of data values

```python
print(f"1st Quartile: {np.percentile(results,25)} below which 25% of the data values lie.")
print(f"2st Quartile: {np.percentile(results,50)} below which 50% of the data values lie.")
print(f"3rd Quartile: {np.percentile(results,75)} below which 75% of the data values lie.")
```

9) **Interquartile Range**: Q3 - Q1\
    IQR is used in outlier detection

```python
q1 = np.percentile(results,25)
q3 = np.percentile(results,75)
iqr = q3 - q1
print(f"Interquartile Range: {iqr}")
```

10) **Outlier Detection**: \
    Industry Standard Rule:\
    Lower Bound --> $Q_1 - 1.5 \times \mathrm{IQR}$\
    Upper Bound --> $Q_3 + 1.5 \times \mathrm{IQR}$\
    Value outside these bounds are potential outliers

```python
arr = np.array([-299,200,300,400,500,1000])
q1 = np.percentile(arr,25)
q3 = np.percentile(arr,75)
iqr = q3 -q1
lower_bound = q1 - 1.5*iqr
upper_bound = q3 + 1.5*iqr
outliers = arr[(arr<lower_bound) | (arr>upper_bound)]
print(outliers)
```

11) **Correlation**: Interdependence between variables

```python
hours_studied = np.array([1,2,3,4,5])
scores = np.array([40,50,60,70,80])
np.corrcoef(hours_studied,scores)
```

Interpretation: Correlation near:\
+1 → strong positive\
 0 → no relationship\
-1 → strong negative
