**<center>Vectorization Mathematics</center>**\
Applying an operation to an entire array at once

**1. Vectorized Arithmetic Operations**\
Let's create array to work on:

```python
import numpy as np
arr = np.array([10,20,30,40])
```

a) Addition (+)

```python
arr + 5
```

b) Subtraction (-)

```python
arr - 5
```

c) Multiplication (*)

```python
arr * 2
```

d) Division (/)

```python
arr / 10
```

e) Exponent (**)

```python
arr ** 2
```

f) Floor Division (//)

```python
arr // 3
```

g) Modulus (%)

```python
arr % 3
```

**4. Universal Functions (ufuncs)**

a) Square Root

```python
arr = np.array([1,4,9,16])
np.sqrt(arr)
```

b) Absolute Value

```python
arr = np.array([-5,-2,4,-1])
np.abs(arr)
```

c) Exponential e

```python
np.exp([1,2,3])
```

d) Natural Logarithm

```python
print(np.log([1]))
print(np.log([np.e]))
print(np.log([np.e**2]))
```

**5. Trigonometric Functions**

```python
print(np.sin(np.pi/6))
print(np.cos(np.pi/6))
print(np.tan(np.pi/3))
```

**6. Aggregation Functions**\
Let's create array first

```python
sales = np.array([100,200,300,400])
```

a) Sum

```python
np.sum(sales)
```

b) Mean

```python
np.mean(sales)
```

c) Min

```python
np.min(sales)
```

d) Max

```python
np.max(sales)
```

e) Standard Deviation

```python
np.std(sales)
```

f) Variance

```python
np.var(sales)
```

**7. Axis wise operations**

```python
sales = np.array([
    [100,200,300],
    [400,500,600],
    [700,800,900]
])
```

```python
print("Normal Sum: ",np.sum(sales))
```

```python
print("Column Wise Sum: ",np.sum(sales,axis=0))
```

```python
print("Row Wise Sum: ",np.sum(sales,axis=1))
```
