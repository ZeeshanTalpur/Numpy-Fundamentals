### Stage 1 — Dataset Understanding
Before analyzing anything, understand the dataset.\
**Questions:**\
How many customers are present?\
How many features exist?\
What is the shape of the dataset?\
Separate every column into its own NumPy array.\
Verify the datatype of the dataset.\
**Deliverable:**\
Dataset Overview Report

```python
rows,columns = data.shape
print(f"Number of customers: {rows}")
print(f"Number of features: {columns}")
print(f"Shape of the dataset: {data.shape}")
cust_id = data[:,0]
age = data[:,1]
salary = data[:,2]
spending_score = data[:,3]
years = data[:,4]
print(f"Customer IDs : {cust_id}")
print(f"Ages: {age}")
print(f"Salaries: {salary}")
print(f"Spending Score: {spending_score}")
print(f"Years as Customer: {years}")
print(f"Datatype of the dataset: {data.dtype}")
```
