### Stage 6 — Outlier Investigation
Finance suspects unusual salaries.\
**Tasks:**\
Detect salary outliers using the IQR method.\
List all detected outlier salaries.\
Identify the customer IDs associated with those outliers.\
Explain whether these customers should be investigated further.\
**Deliverable:**\
Outlier Analysis Report

```python
q1 = np.percentile(data[:,2],25)
q3 = np.percentile(data[:,2],75)
iqr = q3-q1
lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr
outliers = data[(data[:,2]<lower) | (data[:,2]>upper)]
print(f"Outliers: \n{outliers}")
index = np.where((data[:,2]<lower) | (data[:,2]>upper))
print(f"Customer IDs of outliers: \n{data[index,0]}")
```

### Stage 7 — Customer Segmentation
Marketing wants customer categories.\
**Create:**\
Low Income\
Medium Income\
High Income\
You must decide reasonable salary ranges.\
**Tasks:**\
Assign every customer to a segment.\
Count customers in each segment.\
Calculate the percentage in each segment.\
**Deliverable:**\
Income Segmentation Report

```python
salary = data[:,2]
conditions = (
    salary<=60000,
    (salary>60000) & (salary<=100000),
    salary>100000
)
choices = ["Low income", "Medium income", "High income"]
segments = np.select(conditions, choices, default="")
data_segments = np.column_stack((data,segments))
low = np.array(np.where(data_segments[:,5] == "Low income"))
medium = np.array(np.where(data_segments[:,5] == "Medium income"))
high = np.array(np.where(data_segments[:,5] == "High income"))
print(f"Number of low income customers: {low.size} equivalent to {(low.size/(low.size+medium.size+high.size))*100}%")
print(f"Number of medium income customers: {medium.size} equivalent to {(medium.size/(low.size+medium.size+high.size))*100}%")
print(f"Number of high income customers: {high.size} equivalent to {(high.size/(low.size+medium.size+high.size))*100}%")
```

### Stage 9 — Correlation Analysis
Management wants to understand customer behavior.\
**Analyze relationships among:**\
Age\
Salary\
SpendingScore\
YearsAsCustomer\
**Tasks:**\
Generate a correlation matrix.\
Identify strongest positive correlation.\
Identify strongest negative correlation.\
Identify weakest correlation.\
Interpret each finding in business terms.\
**Deliverable:**\
Relationship Analysis Report

```python
matrix = np.corrcoef(data,rowvar=False)
print(f"Correlation Matrix: \n{matrix}")
corr_no_diag = matrix.copy()
np.fill_diagonal(corr_no_diag, -np.inf) # replace diagonal with negative infinity so the max value is not 1
max_corr = np.max(corr_no_diag)
print(f"Maximum Positive Correlation:{max_corr}")
np.fill_diagonal(corr_no_diag, np.inf)
min_corr = np.min(corr_no_diag)
print(f"Strongest Negative Correlation: {min_corr}")
weakest = np.min(np.abs(corr_no_diag))
print(f"Weakest correlation: {weakest}")
```

### Stage 10 — Feature Engineering
Create at least three new features.\

```python
age = data[:,1]
salary = data[:,2]
spending = data[:,3]
years = data[:,4]
value_score = salary * spending
loyalty_score = years * spending
spending_ratio = spending / salary
X = np.column_stack((
    age,
    salary,
    spending,
    years,
    value_score,
    loyalty_score,
    spending_ratio
))
print("Enhanced Feature Matrix:")
print(X)
print("\nShape:", X.shape)
```

### Stage 11 — Normalization and Standardization
Data is being prepared for Machine Learning.\
**Tasks:**\
Normalize all numerical features.\
Standardize all numerical features.\
Verify the normalization results.\
Verify the standardization results.\
**Deliverable:**\

```python
normalized_age = (age-age.min())/(age.max()-age.min())
normalized_salary = (salary-salary.min())/(salary.max()-salary.min())
normalized_spending_score = (spending_score-spending_score.min())/(spending_score.max()-spending_score.min())
normalized_years = (years-years.min())/(years.max()-years.min())
print("After Normalization:")
print(f" Minimum value in ages: {normalized_age.min()} and Maximum value in ages: {normalized_age.max()}")
print(f" Minimum value in salary: {normalized_salary.min()} and Maximum value in salary: {normalized_salary.max()}")
print(f" Minimum value in spending score: {normalized_spending_score.min()} and Maximum value in spending score: {normalized_spending_score.max()}")
print(f" Minimum value in years as customers: {normalized_years.min()} and Maximum value in years as customers: {normalized_years.max()}")
```

```python
standardized_age = (age-np.mean(age))/np.std(age)
standardized_salary = (salary-np.mean(salary))/np.std(salary)
standardized_spending_score =(spending_score-np.mean(spending_score))/np.std(spending_score)
standardized_years = (years-np.mean(years))/np.std(years)
print("After Standardization:")
print(f" Mean of ages: {np.mean(standardized_age)} and Standard Deviation of ages: {np.std(standardized_age)}")
print(f" Mean of salary: {np.mean(standardized_salary)} and  Standard Deviation of salary: {np.std(standardized_salary)}")
print(f" Mean of spending score: {np.mean(standardized_spending_score)} and Standard Deviation of spending score: {np.std(standardized_spending_score)}")
print(f" Mean of years as customer: {np.mean(standardized_years)} and Standard Deviation of years as customers: {np.std(standardized_years)}")
```
