**Array Attributes**


Creating 1D, 2D and 3D arrays to examine their attributes

```python
import numpy as np
arr1d = np.array([2,3,4,5])
arr2d=np.array([
    [1,2,3],
    [2,3,4]
])
arr3d = np.array([
    [
        [1,2,3],
        [4,5,6]
    ],
    [
        [2,4,6],
        [8,10,12]
        ]
])
```

**1. Shape**\
tells us the size of array in each dimension

a. 1D Array:

```python
print("Array: ",arr1d)
print("Shape: ",arr1d.shape)
```

b. 2D Array:

```python
print("Array: \n",arr2d)
print("Shape: ",arr2d.shape)
```

c. 3D Array:

```python
print("Array: \n",arr3d)
print("Shape: ",arr3d.shape)
```

**2. ndim**\
Number of dimensions

```python
print("Dimensions of 1D Array: ",arr1d.ndim)
print("Dimensions of 2D Array: ",arr2d.ndim)
print("Dimensions of 3D Array: ",arr3d.ndim)
```

**3. size**\
Total number of elements in the array

```python
print("Size of 1D Array we created: ",arr1d.size)
print("Size of 2D Array we created: ",arr2d.size)
print("Size of 3D Array we created: ",arr3d.size)
```

Fun Fact:\
Size of an array = Product of shape

```python
print("Shape of 2D Array: ", arr2d.shape)
print("Size of that Array: ", arr2d.size)
print("Shape of 3D Array: ", arr3d.shape)
print("Size of that Array: ", arr3d.size)
```

 **4. dtype**\
 Data type of elements

```python
print("Data type of elements of 1d Array: ", arr1d.dtype)
print("Data type of elements of 2d Array: ", arr2d.dtype)
print("Data type of elements of 3d Array: ", arr3d.dtype)
```

**5. itemsize**\
Memory occupied by one element(in bytes)

```python
print("Size of each element in our 1D Array: ",arr1d.itemsize)
print("Size of each element in our 2D Array: ",arr2d.itemsize)
print("Size of each element in our 3D Array: ",arr3d.itemsize)
```

**6. nbytes**\
Memory occupied by entire array

```python
print("Memory occupied by 1D Array: ",arr1d.nbytes)
print("Memory occupied by 2D Array: ",arr2d.nbytes)
print("Memory occupied by 3D Array: ",arr3d.nbytes)
```

Another fun fact:\
nbytes = size * itemsize

```python
print("Size of our 3D Array: ",arr3d.size)
print("Itemsize of our 3D Array: ",arr3d.itemsize)
print("Memory occupied by entire 3D Array: ",arr3d.nbytes)
```

**Tips**\
How many dimensions? → ndim\
How many rows/columns? → shape\
How many total values? → size\
What type of values? → dtype\
Memory per value? → itemsize\
Total memory? → nbytes
