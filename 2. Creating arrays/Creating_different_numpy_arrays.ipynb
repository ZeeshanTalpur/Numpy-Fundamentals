__1. Creating an array from a list__


a. 1D Array:

```python
import numpy as np
arr = np.array([1,2,3,4,5])
print(arr)
```

b. 2D Array:

```python
arr=np.array([
    [1,2,3],
    [2,4,6],
    [3,6,9]
])
print(arr)
```

__2. Create an array of Zeros__

a. 1D Array:

```python
arr=np.zeros(5)
print(arr)
```

b. 2D Array:

```python
arr=np.zeros((3,3))
print(arr)
```

__3. Creating an array of Ones__

a. 1D Array:

```python
arr=np.ones(5)
print(arr)
```

b. 2D Array:

```python
arr=np.ones((3,2))
print(arr)
```

__4. Creating an Empty Array__

```python
arr=np.empty(5)
print(arr)
```

**5. Creating an Array with a specific value**

a. 1D Array:

```python
arr=np.full(5,68)
print(arr)
```

b. 2D Array:

```python
arr=np.full((2,3),5)
print(arr)
```

**6. Creating a sequence**

a. with max limit only

```python
arr=np.arange(10)
print(arr)
```

b. with both min and max limit

```python
arr=np.arange(1,23)
print(arr)
```

c. with step specified too

```python
arr=np.arange(0,10,2)
print(arr)
```

**7. Creating evenly spaced numbers**

```python
arr=np.linspace(0,10,5)
```

Reference to the above example:\
0 --> Start\
10 --> Stop\
5 --> Number of evenly spaced numbers in this interval

```python
print(arr)
```

**8. Creating an identity matrix**

```python
arr=np.eye(3)
print(arr)
```

**9. Creating random array**

a. random values between 0 and 1

```python
arr=np.random.random(5)
print(arr)
```

b. random integers within a given range

```python
arr=np.random.randint(0,100,5)
```

Reference to the above example:\
0 --> Start\
100 --> Stop\
5 --> Size of array

```python
print(arr)
```
