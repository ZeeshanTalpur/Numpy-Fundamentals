**Workflow 1: Customer Analytics**

```python
import numpy as np
customers = np.array([
    [22,50000,2000],
    [25,70000,3000],
    [30,90000,5000],
    [21,45000,1500],
    [35,120000,7000]
])
```

Step 1: Understand Shape

```python
customers.shape
```

This means 5 customers and 3 features

Step 2: Separate Features

```python
age = customers[:,0]
salary = customers[:,1]
spending = customers[:,2]
print(f"Ages: {age}")
print(f"Salary: {salary}")
print(f"Spending: {spending}")
```

Step 3: Basic Statistics

```python
average_customer = np.array([np.mean(age),np.mean(salary),np.mean(spending)])
print(f"An Average Customer: {average_customer}")
```

Step 4: Spread

```python
print(f"Spread in the salary: {np.std(salary)}")
```

High Standard Deviation --> Customers differ greatly\
Low Standard Deviation --> Customers are similar

Step 5: High Value Customers\
Business question:\
Which customers spend more than average?

```python
customers[spending>np.mean(spending)]
```

This is a classic segmentation task.

**Workflow 2: Outlier Detection**

```python
salary = np.array([45000,50000,52000,55000,60000,250000])
```

Step 1: Compute Quartiles

```python
q1 = np.percentile(salary,25)
q3 = np.percentile(salary,75)
```

Step 2: Interquartile Range

```python
iqr = q3 - q1
```

Step 3: Defining Limits

```python
lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr
```

Step 4: Detect Outliers

```python
salary[(salary<lower)|(salary>upper)]
```

**Workflow 3: Correlation Analysis**

```python
salary = np.array([30000,50000,70000,90000])
spending = np.array([1000,2500,4000,6000])
```

Correlation:

```python
corr = np.corrcoef(salary,spending)
print(f"Correlation Matrix:\n{corr}")
```

Extract:

```python
corr[0,1]
```

Interpretation:\
Strong Positive Relationship\
High Income Customers Spend more.

**Workflow 4: Customer Segmentation**

```python
salary = np.array([30000,50000,70000,100000])
```

Business wants:\
Low Income\
Medium Income\
High Income

```python
conditions = [
    salary<40000,
    (salary>=40000) & (salary<80000),
    salary>80000
]
choices=["low","medium","high"]
np.select(conditions,choices,"")
```

This is customer segmentation

**Workflow 5: Missing Data Investigation**\
Real Datasets are messy

```python
salary = np.array([50000,np.nan,70000,np.nan,90000])
```

How many missing?

```python
np.sum(np.isnan(salary))
```

What percent of values are missing?

```python
(np.sum(np.isnan(salary))/salary.size)*100
```

Mean ignoring NaNs

```python
np.nanmean(salary)
```

**Workflow 6: Product Performance Analysis**

```python
sales = np.array([1200,500,2000,1500,3000])
products = np.array(["A","B","C","D","E"])
```

Best Product: 

```python
idx = np.argmax(sales)
products[idx]
```

Worst Product:

```python
idx = np.argmin(sales)
products[idx]
```

Ranking Products:

```python
products[np.argsort(sales)]
```

**Business usage:**\
Top sellers\
Bottom sellers\
Inventory planning
