**Examining differences between Lists in Python and Arrays in Numpy**

Creating a List:

```python
a = [1,2,3,4]
print(a)
print(type(a))
```

Creating an Array:

```python
import numpy as np
b = np.array([1,2,3,4])
print(b)
print(type(b))
```

**Examining speed difference**

Time taken for operation on a Python List:

```python
import time
a=[1,2,3,4]*100000
start=time.time()
for num in range(len(a)):
    a[num]=a[num]*2
time_taken=time.time()-start
print(time_taken)
```

Time taken for operation on a NumPy Array:

```python
b=np.array(a)
start=time.time()
new_b=b*2
time_taken=time.time()-start
print(time_taken)
```

**Examining the memory difference:**

```python
import sys
import numpy as np
py_list = list(range(1000000))
np_array = np.array(py_list)
list_memory = sys.getsizeof(py_list) #Size of just the list container
for item in py_list:
    list_memory += sys.getsizeof(item) #Size of the individual elements added
array_memory = np_array.nbytes #Size of array
print(f"Python List Memory: {list_memory / (1024**2):.2f} MB")
print(f"NumPy Array Memory: {array_memory / (1024**2):.2f} MB")
```

**Final Verdict:**



A Python List is created without the need of any module while a NumPy Array requires NumPy library.
A Python List works almost 100 times slower as compared to a NumPy Array.
A Python List needs loops for mathematical operations whereas a NumPy Array undergoes vectorized operations.
A Python List requires more than 4 times greater memory than a NumPy Array for storage.
