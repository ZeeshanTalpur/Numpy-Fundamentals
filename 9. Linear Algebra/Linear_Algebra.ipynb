**Part 4: Matrix Multiplication:**

A * B returns element wise multiplication.\
For matrix multiplication,\
A @ B\
Or \
np.matmul(A,B)

````python
A = np.array([
    [1,2],
    [3,4]
])

B = np.array([
    [5,6],
    [7,8]
])

A @ B
````

or

````python
np.matmul(A,B)
````

**Insight:** Every element i.e., A[i][j] in the Product Matrix comes from the dot product of ith row of first matrix and jth column of second matrix.\
**Shape Rule:** While multiplying,\
Matrix A --> shape=(m,n)\
Matrix B --> shape=(n,p)\
The shape of product will be (m,p).\
**Eligibility Rule:** Multiplication is only possible if number of columns of first and rows of second are equal.

**Part 5: Tranpose in Linear Algebra**\
swapping of rows and columns

````python
A = np.array([
    [1,2,3],
    [4,5,6]
])
print(f"Original Matrix:\n{A}")
print(f"Shape of Original Array: {A.shape}")
B = A.T
print(f"Transpose:\n{B}")
print(f"Shape of Transpose: {B.shape}")
````

**Part 6: Identity Matrix**\
Identity Matrix behaves like number 1.

````python
I = np.eye(3)
print(f"Identity Matrix:\n{I}")
````

A @ I = A, where A is a matrix of same order as I

````python
print(f"Matrix A:\n{A}")
print(f"A @ I:\n{A @ I}")
````

**Part 7: Matrix Inverse**\
Like numbers, matrices have inverses too\
A @ inverse of A = I

````python
print(f"Matrix A:\n{A}")
print(f"Inverse of A:\n{np.linalg.inv(A)}")
print(f"A @ Inverse of A:\n{np.round(A @ np.linalg.inv(A))}")
````

**Part 8: Determinant**\
A number associated with matrix which tells if a matrix is singular(not invertible) or non-singular(invertible)\
A matrix is singular if the determinant is equal to 0.\
A matrix is non-singular if the determinant is not equal to 0.

````python
np.linalg.det(A)
````

In the above example, the determinant is not equal to 0, which means the given matrix is non-singular and hence, invertible.

**Part 9: Solving System of Equation**\
**Suppose:**\
x + y = 5\
2x + y = 8\
**In matrix form,**\
A = coefficient matrix(a matrix that consists of all the coefficients given with variables in the system of equations)\
b = constant matrix(a matrix that contains all the constants on the right hand sides of system of equations)\
**To solve this system of equations for values of x and y**\
np.linalg.solve(A,b)

````python
A = np.array([
    [1,1],
    [2,1]
])

b = np.array([
    5,
    8
])
np.linalg.solve(A,b)
````

**Output of the above code implies that:**\
x = 3\
y = 2

**Part 10: Norms**\
Norm is simply the length(magnitude) of a vector\
**What Does "Length" Mean?**\
Suppose you're standing at:\
(0,0)\
and walk to:\
(3,4)\
**Mathematically:**\
$d = \sqrt{(3-0)^2 + (4-0)^2} = 5$\
**Graphically**:
```text
      (3,4)
        *
       /|
      / |
     /  |
    /   |
   /    |
  *-----*
(0,0)  (3,0)
```
The horizontal movement is: 3\
The vertical movement is: 4\
The actual distance travelled is: 5\
While calculating the norm(length of a vector), remember that a vector always starts from (0,0).\
Hence, the formula becomes\
$\|\mathbf{v}\| = \sqrt{x^2 + y^2}$

````python
v = np.array([3,4])
np.linalg.norm(v)
````
